# Gradient Descent Simulation Exercise
## Objective
Understand the principles of gradient descent by manually computing and applying gradient descent steps to a simple function.

## Background
Gradient descent is an optimization algorithm used to minimize a function by iteratively moving towards the minimum value of the function.

## Simple Function
We'll use a simple quadratic function for this exercise: f(x)=x^2

## Gradient Descent Steps
1. Start with an initial guess for x. 
2. Compute the gradient (derivative) of f at this point.
3. Update x by moving in the direction opposite to the gradient.
4. Repeat until convergence.


# Setup
Import necessary libraries and define the function and its derivative.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x ** 2

def df(x):
    return 2 * x
```

# Simulation
Manually perform gradient descent on the function f(x).

```python
# Gradient Descent Parameters
x = 10  # initial guess
learning_rate = 0.1
iterations = 25

# Lists to store x values and f(x) values
x_values, f_values = [], []

# Gradient Descent Loop
for i in range(iterations):
    x_values.append(x)
    f_values.append(f(x))

    # Calculate the gradient
    grad = df(x)

    # Update x
    x = x - learning_rate * grad

# Plotting the function and the steps
plt.plot(np.linspace(-10, 10, 100), f(np.linspace(-10, 10, 100)), label='f(x) = x^2')
plt.scatter(x_values, f_values, color='red')
plt.title('Gradient Descent on f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Varying Learning Rate

```python
lr = [0.01, 0.1, 0.5, 1]
plt.figure(figsize=(15,6))

for learning_rate in lr:

    # Gradient Descent Parameters
    x = 10  # initial guess
    iterations = 25

    # Lists to store x values and f(x) values
    x_values, f_values = [], []

# Gradient Descent Loop
    for i in range(iterations):
        x_values.append(x)
        f_values.append(f(x))

        # Calculate the gradient
        grad = df(x)

        # Update x
        x = x - learning_rate * grad

    # Plotting the function and the steps
    
    plt.subplot(1,4,lr.index(learning_rate) + 1)
    plt.plot(np.linspace(-10, 10, 100), f(np.linspace(-10, 10, 100)), label='f(x) = x^2')
    plt.scatter(x_values, f_values, color='red')
    plt.title(f' Learning Rate = {learning_rate}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
plt.tight_layout()
plt.show()
```

## Varying Starting Point

```python
# Different starting points
starting_points = [2, 5, -5, 10]
plt.figure(figsize=(15, 6))

for sp in starting_points:

    # Gradient Descent Parameters
    x = sp  # initial guess
    iterations = 25

    # Lists to store x values and f(x) values
    x_values, f_values = [], []

# Gradient Descent Loop
    for i in range(iterations):
        x_values.append(x)
        f_values.append(f(x))

        # Calculate the gradient
        grad = df(x)

        # Update x
        x = x - learning_rate * grad

    # Plotting the function and the steps
    
    plt.subplot(1,4,starting_points.index(sp) + 1)
    plt.plot(np.linspace(-10, 10, 100), f(np.linspace(-10, 10, 100)), label='f(x) = x^2')
    plt.scatter(x_values, f_values, color='red')
    plt.title(f' Learning Rate = {learning_rate}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
plt.tight_layout()
plt.show()
```
